WUP
====


```python
import seaborn as sns
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

df = pd.read_csv("wup2.tsv", delimiter="\t")

sns.set()
sns.set_context("talk")                                    
plot = sns.lineplot(x="vsize", y="human_dynamic", hue="corpus", style="corpus", markers=["o","X","P","D","s"], data=df)
plt.xlabel("Vector size")
plt.ylabel(r"Spearman's correlation")
plt.title("WordNet graph (noun synsets)")

# removing legend title
plt.legend(loc='best')
handles, labels = plot.get_legend_handles_labels()
plot.legend(handles=handles[1:], labels=labels[1:])

plot.figure.savefig("wup.pdf", bbox_inches='tight')
```

FB15k
=====

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("fb.tsv", delimiter="\t")

sns.set()
sns.set_context("talk")                                    

plot = sns.lineplot(x="vsize", y="Pearson", hue="modelname", data=df, style="modelname", markers=["o","X","P","D"])
plot.legend_.remove()

plt.title("Freebase graph (FB15k-237)")
plt.xlabel("Vector size")
plt.ylabel("Pearson's correlation")
plot.figure.savefig("fb15.pdf", bbox_inches='tight')

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("dbpedia.tsv", delimiter="\t")

sns.set()
sns.set_context("talk")                                    

plot = sns.lineplot(x="vsize", y="Pearson", hue="modelname", data=df, style="modelname", markers=["o","X","P","D","."])
plot.legend_.remove()

plt.title("DBpedia graph (DB100k)")
plt.xlabel("Vector size")
plt.ylabel("Pearson's correlation")
plot.figure.savefig("dbpedia.pdf", bbox_inches='tight')
```
